## Q2. How does broadcasting work in NumPy?
### Ans : Broadcasting in NumPy is a powerful mechanism that allows arithmetic operations to be performed on arrays of different shapes and sizes efficiently. Instead of requiring arrays to have the exact same shape, NumPy "broadcasts" the smaller array across the larger one so that their shapes become compatible for element-wise operations without actually copying data. This leads to faster computation and lower memory usage.

>> ### The broadcasting rules work as follows :

>1 .  If arrays have different numbers of dimensions, the shape of the smaller array is padded with ones on its leading (left) side.

>2 . Dimensions are compatible if they are equal or if one of them is 1.

>3 . If the size in any dimension does not match and neither equals 1, a ValueError is raised.

>4 . Arrays with dimensions of size 1 are "stretched" or broadcast to match the other array's size in that dimension.

>> **For example**, adding a 1D array of shape (3,) to a 2D array of shape (3, 3) results in the 1D array being broadcast to behave like it has shape (3, 3) by replicating its values along the added dimension. This enables element-wise addition without explicit looping or reshaping.

> Broadcasting thus enables vectorized operations on differently shaped arrays, improving code simplicity, efficiency, and speed

## Q6. What are the differences between NumPy arrays and Python lists?

### Ans :  NumPy arrays and Python lists are both widely used in Python for storing collections of data, but they differ greatly in terms of structure, performance, and flexibility.

1. Storage and Speed

>- NumPy arrays store data in a single block of memory and all elements are of the same data type.

> → This makes them much faster and more memory efficient.

>- Python lists can hold different types (like integers, strings, floats) and are stored as pointers to objects.

> → This flexibility makes them slower and less efficient for numerical work.

2. Data Type

>- NumPy: All elements must have the same type (e.g., all floats or all ints).

>- Lists: Each element can be of a different type.

3. Operations

>- NumPy arrays support element-wise operations.

4. Functionality

>- NumPy arrays come with tons of mathematical functions (np.mean, np.sum, np.dot, etc.) built in.

>- Lists don't have direct math operations — you'd need loops or list comprehensions.

5. Size and Shape

>- NumPy supports multi-dimensional arrays (like matrices).

>- Lists can be nested, but they don't behave like true multi-dimensional arrays — they're just lists inside lists.

6. Use Case

>- Use NumPy arrays for numerical and scientific computing (data analysis, machine learning, etc.).

>- Use lists for general-purpose storage or when your data isn’t strictly numerical.

## Q8. What does the term “vectorized operation” mean in NumPy?

### Ans : A "vectorized operation" in NumPy refers to performing mathematical or logical operations on entire arrays at once, without needing explicit loops to iterate through each element. Instead of applying an operation element-by-element via Python loops, vectorized operations leverage highly optimized, precompiled C code behind the scenes to perform element-wise calculations on arrays efficiently and quickly.

>**For example**, adding a scalar number to every element of a NumPy array or adding two arrays element-wise is done in a vectorized manner, resulting in cleaner code and significant performance improvements compared to manual iteration. Vectorization also typically supports broadcasting, which allows operations on arrays of different shapes and sizes.

> ### This process leads to faster execution because :

>- It eliminates explicit Python-level loops.

>- It minimizes type-checking overhead by operating on homogeneous data types.

>- It utilizes low-level optimizations inherent in NumPy's C implementation.

## Q21. Why is NumPy’s array slicing faster than Python’s list slicing?

### Ans NumPy's array slicing is faster than Python's list slicing mainly because of how data is stored and handled in memory.

>> ## 1. Memory layout

>- NumPy arrays store elements in a contiguous block of memory—all values are of the same type (like all integers or floats).

>- Python lists store pointers to objects scattered across memory, which takes more time to access.

> When NumPy slices an array, it doesn't copy data; it just creates a view (a window) into the same memory block. That's much faster.

>> ## 2. Views vs. Copies

>- NumPy slicing → View: Doesn’t copy data, just references it.

>- List slicing → Copy: Creates a completely new list and copies each element.

> That copying step is what slows down list slicing.

>> ## 3.  Low-level (C-based) optimization

> NumPy is written in C, so its operations use highly optimized C loops that work directly with memory addresses.
Python lists rely on the interpreter, which adds overhead for every element.

>> ## 4. Homogeneous vs. Heterogeneous data

> NumPy arrays contain one data type, so they don’t need to check the type of each element during slicing.
Python lists can hold mixed types, so Python does extra work managing those references.

## Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Maths': [88, 92, 79, 93, 85],
    'Science': [90, 85, 88, 95, 80],
    'English': [70, 78, 82, 72, 85],
    'History': [60, 65, 70, 63, 75]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap to visualize it
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Add a title
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Faisal', 'Sameer', 'Karan', 'Aditi', 'Rohit'],
    'Marks': [72, 88, 65, 90, 78]
}

df = pd.DataFrame(data)

# Set a threshold
threshold = 75

# Filter rows where Marks are greater than threshold
filtered_df = df[df['Marks'] > threshold]

print("Original DataFrame:")
print(df)

print("\nFiltered DataFrame (Marks > 75):")
print(filtered_df)
```

## Q10. Create a histogram using Seaborn to visualize a distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a sample dataset
data = pd.DataFrame({
    'Age': np.random.randint(18, 60, size=100)
})

# Create a histogram
sns.histplot(data['Age'], bins=10, kde=True, color='skyblue')

# Add labels and title
plt.title('Age Distribution of Participants')
plt.xlabel('Age')
plt.ylabel('Count')

# Display the plot
plt.show()
```

## Q11. Perform matrix multiplication using NumPy.

```python
import numpy as np

# Create two matrices
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Perform matrix multiplication
result = np.dot(A, B)

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nResult of Matrix Multiplication:")
print(result)
```

## Q13. Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import pandas as pd

# Create sample data
data = {
    'X': [2, 4, 6, 8, 10],
    'Y': [1, 3, 5, 7, 9],
    'Z': [10, 15, 13, 17, 20],
    'Category': ['A', 'B', 'A', 'B', 'A']
}

df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = px.scatter_3d(
    df,
    x='X',
    y='Y',
    z='Z',
    color='Category',
    size='Z',
    title='3D Scatter Plot Example'
)

# Show the plot
fig.show()
```
